# Seaborn 그래프

## Scatterplot

0 ~ 1 사이의 임의의 랜덤한 값을 생성합니다. 

```python
np.random.rand(50)
```

0 부터 50개의 값을 순차적으로 생성합니다.

```python
np.arange(50)
```

### x, y, colors, area 설정

* colors 는 임의 값을 color 값으로 변환합니다.
* area는 점의 넓이를 나타냅니다. 값이 커지면 당연히 넓이도 커집니다.

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.arange(50)
area = x * y * 250
```

```python
plt.scatter(x, y, s=area, c=colors)
plt.show()
```

* seaborn 에서는 `size`와 `sizes`를 동시에 지정해줍니다.
* `sizes` 옵션에서는 사이즈의 min, max를 명시해 줍니다.
* `hue`는 컬러 옵션입니다.
* `palette`를 통해 seaborn이 제공하는 아름다운 palette 를 이용하시면 됩니다.

```python
sns.scatterplot(x=x, y=y, size=area, sizes=(area.min(), area.max()), hue=area, palette='icefire')
plt.show()
```

### cmap과 alpha 설정

* cmap에 컬러를 지정하면, 컬러 값을 모두 같게 가져갈 수도 있습니다.
* alpha값은 투명도를 나타내며 0 ~ 1 사이의 값을 지정해 줄 수 있으며, 0에 가까울 수록 투명한 값을 가집니다.

```python
plt.figure(figsize=(15, 5))

plt.subplot(131)
g = sns.scatterplot(x=x, y=y, size=area, sizes=(area.min(), area.max()), color='purple', alpha=0.1)
g.get_legend().remove()
plt.title('alpha=0.1')

plt.subplot(132)
plt.title('alpha=0.5')
g = sns.scatterplot(x=x, y=y, size=area, sizes=(area.min(), area.max()), color='purple', alpha=0.5)
g.get_legend().remove()

plt.subplot(133)
plt.title('alpha=1.0')
g = sns.scatterplot(x=x, y=y, size=area, sizes=(area.min(), area.max()), color='purple', alpha=0.9)
g.get_legend().remove()

plt.show()
```

## Barplot, Barhplot

### Batplot에서 비교 그래프 그리기

matplotlib 을 활용하여 비교 그래프를 그려야 하는 경우

```python
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
x = np.arange(len(x_label))
y_1 = [66, 80, 60, 50, 80, 10]
y_2 = [55, 90, 40, 60, 70, 20]

# 넓이 지정
width = 0.35

# subplots 생성
fig, axes = plt.subplots()

# 넓이 설정
axes.bar(x - width/2, y_1, width, align='center', alpha=0.5)
axes.bar(x + width/2, y_2, width, align='center', alpha=0.8)

# xtick 설정
plt.xticks(x)
axes.set_xticklabels(x_label)
plt.ylabel('Number of Students')
plt.title('Scores')

plt.legend(['john', 'peter'])

plt.show()
```

Seaborn에서는 위의 `matplotlib`과 조금 다른 방식을 취합니다.

**실전 tip.**
* 그래프를 임의로 그려야 하는 경우 -> `matplotlib`
* DataFrame을 가지고 그리는 경우 -> `seaborn`

seaborn 에서는 `hue` 옵션으로 매우 쉽게 비교 **barplot**을 그릴 수 있습니다.

```python
# 샘플 데이터셋 로드
titanic = sns.load_dataset('titanic')
```

```python
sns.barplot(x='sex', y='survived', hue='pclass', data=titanic, palette="muted")
plt.show()
```

## Line Plot

```python
# 샘플 데이터 생성
x = np.arange(0, 10, 0.1)
y = 1 + np.sin(x)
```

```python
# grid 스타일을 설정할 수 있습니다. 
# whitegrid, darkgrid, white, dark, ticks
sns.set_style("darkgrid")

sns.lineplot(x=x, y=y)

plt.xlabel('x value', fontsize=12)
plt.ylabel('y value', fontsize=12)
plt.title('sin graph', fontsize=15)

plt.show()
```

### 2개 이상의 그래프 그리기

* color: 컬러 옵션
* alpha: 투명도 옵션

```python
# 샘플 데이터셋 생성
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

sns.lineplot(x=x, y=y_1, label='1+sin', color='navy', alpha=0.3)
sns.lineplot(x=x, y=y_2, label='1+cos', color='red', alpha=0.7)

plt.xlabel('x value', fontsize=12)
plt.ylabel('y value', fontsize=12)
plt.title('sin and cos graph', fontsize=15)

plt.grid()
plt.legend()

plt.show()
```

### 마커 스타일링

* marker: 마커 옵션

```python
sns.set_style("white")

x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

sns.lineplot(x=x, y=y_1, label='1+sin', color='blue', alpha=0.3, marker='o')
sns.lineplot(x=x, y=y_2, label='1+cos', color='red', alpha=0.7, marker='+')

plt.xlabel('x value', fontsize=15)
plt.ylabel('y value', fontsize=15)
plt.title('sin and cos graph', fontsize=18)

plt.grid()
plt.legend()

plt.show()
```

### 라인 스타일 변경하기

* linestyle: 라인 스타일 변경 옵션

```python
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

sns.lineplot(x=x, y=y_1, label='1+sin', color='blue', linestyle=':')
sns.lineplot(x=x, y=y_2, label='1+cos', color='red', linestyle='-.')

plt.xlabel('x value', fontsize=15)
plt.ylabel('y value', fontsize=15)
plt.title('sin and cos graph', fontsize=18)

plt.grid()
plt.legend()

plt.show()
```

## Histogram

```python
N = 100000
bins = 30

x = np.random.randn(N)

sns.histplot(x, bins=bins, kde=False, color='g')
# 구 버전
# sns.distplot(x, bins=bins, kde=False, color='g')

plt.show()
```

`kde`을 **True**로 설정해주면, **Density**가 Y축에 표기 됩니다.

```python
sns.histplot(x, bins=bins, kde=True, color='g')
# 구 버전
# sns.distplot(x, bins=bins, kde=True, color='g')
```

```python
sns.histplot(y=x, bins=bins, kde=True, color='navy')
# 구 버전
# sns.distplot(x, bins=bins, kde=True, color='navy')
```

* sharey: y축을 다중 그래프가 share
* tight_layout: graph의 패딩을 자동으로 조절해주어 fit한 graph를 생성

### 분포표, 누적분포표

```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axes = plt.subplots(1, 2, tight_layout=True)
fig.set_size_inches(12, 4)

sns.histplot(x=x, bins=bins, ax=axes[0])
sns.histplot(x=x, bins=bins, cumulative=True, ax=axes[1])

# 구버전
# sns.displot(x=x, bins=bins)
# sns.displot(x=x, bins=bins, cumulative=True)

plt.show()
```

## Box Plot

샘플 데이터를 생성합니다.

```python
# 샘플 데이터 생성
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(121)
sns.boxplot(data=data, orient='h', width=0.2)
plt.subplot(122)
sns.boxplot(data=data, orient='v', width=0.2)
plt.show()
```

seaborn에서 boxplot을 그릴 때는DataFrame을 가지고 그릴 때 주로 활용합니다. 

**barplot**과 마찬가지로, 용도에 따라 적절한 라이브러리를 사용합니다.

**실전 tip.**
* 그래프를 임의로 그려야 하는 경우 -> `matplotlib`
* DataFrame을 가지고 그리는 경우 -> `seaborn`

seaborn 에서는 `hue` 옵션으로 매우 쉽게 비교 **boxplot**을 그릴 수 있습니다.

```python
outlier_marker = dict(markerfacecolor='r', marker='o')
sns.boxplot(x='pclass', y='age', hue='survived', data=titanic, flierprops=outlier_marker)
plt.show()
```
